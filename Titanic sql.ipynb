```python
import pandas as pd
import pandasql as ps
```

```python
titanic_disaster_dataset =pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
print(titanic_disaster_dataset.shape,titanic_disaster_dataset[['PassengerId']].drop_duplicates().shape)
titanic_disaster_dataset.head()
```

```python
ps.sqldf("""SELECT * FROM titanic_disaster_dataset""")
```

```python
ps.sqldf("""select df.* from titanic_disaster_dataset AS df""")
```

#### Putting limits in data 

```python
ps.sqldf("""select df.* from titanic_disaster_dataset AS df limit 5""")
```

#### selecting particular table

```python
ps.sqldf("""select df.PassengerId, df.Name from titanic_disaster_dataset as df""")
```

```python
ps.sqldf("""select df.PassengerID as ID_of_Passanger,df.Name as Name_of_Passanger from titanic_disaster_dataset as df""")
```

#### Where condition

```python
ps.sqldf("""select df.*from titanic_disaster_dataset as df WHERE df.age < 1""")
```

```python
ps.sqldf("""select df.*from titanic_disaster_dataset as df where df.age is Null""")
```

```python
ps.sqldf("""select df.*from titanic_disaster_dataset as df where df.age Between 5 And 10""")
```

```python
ps.sqldf("""select df.*from titanic_disaster_dataset as df where df.age < 1 and df.sex = "male"   """)
```

```python
ps.sqldf("""select df.*from titanic_disaster_dataset as df where df.age IN ('65','70','71') """)
```

### To find the perticular name in the data with some specification 
#### 1)for name starting with "ABC"  .Name Like "ABC%"
#### 2)for name ending with "ABC"  .Name Like "%ABC"
#### 3)for name which contains "ABC"  .Name Like "%ABC%"

```python
ps.sqldf("""select df.*from titanic_disaster_dataset as df where df.Name Like "%Mrs.%" """)
```

```python
ps.sqldf("""select df.age, 
df.age * df.age as age_sq, 
df.age+1 as age_plus_one 
from titanic_disaster_dataset as df""")
```

```python
ps.sqldf("""select df.sex,df.ticket,
            df.sex || df.ticket as sex_plus_ticket
            from titanic_disaster_dataset as df""")
```

```python
ps.sqldf("""select df.passengerid,df.name,df.age,
            CASE When df.age is null then '0. missing'
                when df.age <18 then '1. 1-17' 
                when df.age <60 then '2. 18-60' 
                else '3. 60+'END As age_bucket
                From titanic_disaster_dataset as df """)
```

```python
ps.sqldf("""select df.Embarked,
            AVG(df.age) as avg_age,
            sum(df.age) as sum_age,
            MAX(df.age) as max_age,
            MIN(df.age) as min_age,
            COUNT(df.PassengerId) As tot_passengers,
            COUNT(DISTINCT df.ticket) as dist_tickets
            from titanic_disaster_dataset as df
            Group by df.Embarked
            """)
```

```python
ps.sqldf("""select df.Embarked,df.sex,
            AVG(df.age) as avg_age,
            sum(df.age) as sum_age,
            MAX(df.age) as max_age,
            MIN(df.age) as min_age,
            COUNT(df.PassengerId) As tot_passengers,
            COUNT(DISTINCT df.ticket) as dist_tickets
            from titanic_disaster_dataset as df
            Group by df.Embarked,df.sex
            """)
```

```python
ps.sqldf("""select df.Embarked,df.sex,
            AVG(df.age) as avg_age,
            sum(df.age) as sum_age,
            MAX(df.age) as max_age,
            MIN(df.age) as min_age,
            COUNT(df.PassengerId) As tot_passengers,
            COUNT(DISTINCT df.ticket) as dist_tickets
            from titanic_disaster_dataset as df
            where age>=10 and age<=60
            Group by df.Embarked,df.sex
            """)
```

```python
ps.sqldf("""select df.Embarked,df.sex,
            AVG(df.age) as avg_age,
            sum(df.age) as sum_age,
            MAX(df.age) as max_age,
            MIN(df.age) as min_age,
            COUNT(df.PassengerId) As tot_passengers,
            COUNT(DISTINCT df.ticket) as dist_tickets
            from titanic_disaster_dataset as df
            where df.age>=10 and df.age<=60
          
            Group by df.Embarked,df.sex
            Having avg_age >30
            order by df.Embarked desc,
            df.sex desc
            """)
```

```python
cities =pd.DataFrame({'code':['S','C','L'],'city':['Southampotn','Cherbourg','London']})
cities
```

```python
ps.sqldf("""select 
            c.city,
            df.Embarked,df.sex,
            AVG(df.age) as avg_age,
            sum(df.age) as sum_age,
            MAX(df.age) as max_age,
            MIN(df.age) as min_age,
            COUNT(df.PassengerId) As tot_passengers,
            COUNT(DISTINCT df.ticket) as dist_tickets
            from titanic_disaster_dataset as df
            Inner Join cities as c
            On df.Embarked=c.code
            Group by df.Embarked,df.sex
            
            """)
```

```python
ps.sqldf("""select 
            c.city,
            df.Embarked,df.sex,
            AVG(df.age) as avg_age,
            sum(df.age) as sum_age,
            MAX(df.age) as max_age,
            MIN(df.age) as min_age,
            COUNT(df.PassengerId) As tot_passengers,
            COUNT(DISTINCT df.ticket) as dist_tickets
            from titanic_disaster_dataset as df
            Left Join cities as c
            On df.Embarked=c.code
            Group by df.Embarked,df.sex
            
            """)
```
